```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
from matplotlib.ticker import MultipleLocator

#arquivos = ['000127_bancada_1_parafuso', 'bancada_08_1700rpm', 'pat_708991_3460rpm', 'pat_755050_motor']
#arquivos = ['02_000127_1700rpm', '02_bancada_08_1700rpm', '02_708991_3460rpm', '02_755050_1700rpm', '02_755050_1700rpm_parte_de_tras']
arquivos = ['60_frente', '60_tras']
dirDados = 'dadosgil'
dirGraficos = 'graficosgil'
```

```python
def geragraficos(dfa, eixo, cor, posicao, intervalo_grade, arquivo):
    fs = 1000 #frequencia de amostragem (Hz)
    signal = dfa.values
    signal = signal - np.mean(signal)
    
    # Aplicar janela de Hann para reduzir vazamento espectral
    window = np.hanning(len(signal))
    signal_windowed = signal * window
    
    # Calcular FFT
    N = len(signal)
    yf = np.fft.rfft(signal_windowed)  # só metade (frequências positivas)
    xf = np.fft.rfftfreq(N, 1 / fs)
    
    # Calcular magnitude (em escala linear ou logarítmica)
    magnitude = np.abs(yf)
    
    # Plotar
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.plot(xf, magnitude, color=cor, linewidth=1.5)
    plt.title(f'{arquivo}: Espectro de Frequência - Eixo {eixo} ({posicao})', fontsize=14, fontweight='bold')
    ax.set_xlabel('Frequência (Hz)', fontsize=12)
    ax.set_ylabel('Magnitude', fontsize=12)
    ax.set_xlim(0, fs/2)
    
    ax.margins(0)
    max_x = fs / 2
    ticks = np.arange(0, max_x, intervalo_grade)
    ax.set_xticks(ticks)
    ax.grid(True, axis='x', linestyle='-', linewidth=0.8, color='gray', alpha=0.5)
    ax.grid(True, axis='y', linestyle='-', linewidth=0.8, color='gray', alpha=0.5)
    plt.tight_layout()
    return plt
```

```python
for arquivo in arquivos:
    df = pd.read_csv(
        f'{dirDados}/{arquivo}.csv', 
        delimiter=';',
        decimal='.',
        #encoding='utf-8', #latin-1
        #skipinitialspace=True,
        header=None, 
        names=['acelerometro_x', 'acelerometro_y', 'acelerometro_z']
    )
    
    with PdfPages(f'{dirGraficos}/{arquivo}.pdf') as pdf:
        frequencia = 60
        plt = geragraficos(df['acelerometro_x'], 'x', 'blue', 'Axial', frequencia, arquivo)
        pdf.savefig()  
        plt.close()    
    
        plt = geragraficos(df['acelerometro_y'], 'y', 'red', 'Vertical', frequencia, arquivo)
        pdf.savefig()  
        plt.close()   
    
        plt = geragraficos(df['acelerometro_z'], 'z', 'limegreen', 'Horizontal', frequencia, arquivo)
        pdf.savefig()  
        plt.close()   
```

```python
def plot_boxplot_com_estatisticas(df, colunas, titulo='Boxplot'):
    fig, ax = plt.subplots(figsize=(12, 6))
    
    # Criar boxplot
    bp = ax.boxplot([df[col].dropna() for col in colunas],
                    labels=colunas,
                    patch_artist=True,
                    showmeans=True,
                    meanline=True,
                    showfliers=True,
                    whiskerprops=dict(linewidth=1.5),
                    boxprops=dict(linewidth=1.5),
                    medianprops=dict(color='red', linewidth=2),
                    meanprops=dict(color='green', linewidth=2, linestyle='--'))
    
    # Colorir as caixas
    colors = ['lightblue', 'lightgreen', 'lightcoral']
    for patch, color in zip(bp['boxes'], colors[:len(colunas)]):
        patch.set_facecolor(color)
        patch.set_alpha(0.7)
    
    # Adicionar grid
    ax.grid(True, axis='y', alpha=0.3, linestyle='--')
    
    # Título e labels
    ax.set_title(titulo, fontsize=14, fontweight='bold')
    ax.set_ylabel('Amplitude', fontsize=12)
    ax.set_xlabel('Eixos', fontsize=12)
    
    # Adicionar legenda para média e mediana
    ax.plot([], [], color='red', linewidth=2, label='Mediana')
    ax.plot([], [], color='green', linewidth=2, linestyle='--', label='Média')
    ax.legend(loc='upper right')
    
    plt.tight_layout()
    return fig, ax

# Uso
fig, ax = plot_boxplot_com_estatisticas(
    df, 
    ['acelerometro_x', 'acelerometro_y', 'acelerometro_z'],
    titulo='Análise de Vibração - Boxplot por Eixo'
)
plt.show()
```

```python
import numpy as np
import pandas as pd
#arquivos = ['000127_bancada_1_parafuso', 'bancada_08_1700rpm', 'pat_708991_3460rpm', 'pat_755050_motor']
arquivos = ['02_000127_1700rpm', '02_bancada_08_1700rpm', '02_708991_3460rpm', '02_755050_1700rpm', '02_755050_1700rpm_parte_de_tras', '000127_bancada_1_parafuso']
#arquivos = ['02_000127_1700rpm', '02_bancada_08_1700rpm']
#arquivo = '02_000127_1700rpm'
#df = pd.read_csv(f'dados/duplicados/{arquivo}.csv')
```

```python
def calculate_std(signal: np.ndarray, ddof: int = 1) -> float:
    """Sample Standard Deviation - STD"""
    return np.std(signal, ddof=ddof)

def calculate_rms(signal: np.ndarray) -> float:
    """Root Mean Square - RMS"""
    return np.sqrt(np.mean(np.square(signal)))

def calculate_mav(signal: np.ndarray) -> float:
    """Mean Absolute Value - MAV"""
    return np.mean(np.abs(signal))

def calculate_wfl(signal: np.ndarray) -> float:
    """Waveform Length (WL ou WFL)"""
    return np.sum(np.abs(np.diff(signal)))
```

```python
def extract_features_windowed(df: pd.DataFrame, 
                              column: str, 
                              window_size: int = 200, 
                              step: int = 50) -> pd.DataFrame:
    
    signal = df[column].values
    results = []
    
    for start in range(0, len(signal) - window_size + 1, step):
        window = signal[start:start + window_size]
        results.append({
            #'inicio': start,
            #'fim': start + window_size,
            f'{column}_STD': calculate_std(window),
            f'{column}_RMS': calculate_rms(window),
            f'{column}_MAV': calculate_mav(window),
            f'{column}_WFL': calculate_wfl(window)
        })
    
    return pd.DataFrame(results)

#df_arquivos = pd.DataFrame(columns=['x', 'y', 'z', 'label'])
df_arquivos = pd.DataFrame()
for arquivo in arquivos:
    df = pd.read_csv(
        f'dados/duplicados/{arquivo}.csv', 
        #delimiter=';',
        #decimal=',',
        #encoding='utf-8', #latin-1
        #skipinitialspace=True,
        #header=None, 
        names=['x', 'y', 'z']
    )
    
    features_dfx = extract_features_windowed(df, 'x')
    features_dfy = extract_features_windowed(df, 'y')
    features_dfz = extract_features_windowed(df, 'z')

    df_arquivo = pd.concat([features_dfx, features_dfy, features_dfz], axis=1)
    
    if arquivo == '02_000127_1700rpm':
        df_arquivo['label'] = 'saudavel'
    elif arquivo == '02_bancada_08_1700rpm':
        df_arquivo['label'] = 'rolamento'
    elif arquivo == '02_708991_3460rpm':
        df_arquivo['label'] = 'estator'
    elif arquivo == '02_755050_1700rpm':
        df_arquivo['label'] = 'desalinhamento'
    elif arquivo == '02_755050_1700rpm_parte_de_tras':
        df_arquivo['label'] = 'desalinhamento'
    elif arquivo == '000127_bancada_1_parafuso':
        df_arquivo['label'] = 'desbalanceamento'

    df_arquivos = pd.concat([df_arquivos, df_arquivo], ignore_index=True)

print(df_arquivos)
df_arquivos.to_csv('df_arquivos.csv', index=False)
```
